```python
# 1. Drop the junk column
df = df.drop(columns=['Unnamed: 22'])

# 2. Fill fulfilled-by with "Unknown"
df['fulfilled-by'] = df['fulfilled-by'].fillna('Unknown')

# 3. Fill promotion-ids with "No Promotion"
df['promotion-ids'] = df['promotion-ids'].fillna('No Promotion')

# 4. Fill Courier Status with "Unknown"
df['Courier Status'] = df['Courier Status'].fillna('Unknown')

# 5. Handle currency and Amount (same 7795 nulls - likely cancelled orders)
df['Amount'] = df['Amount'].fillna(0)
df['currency'] = df['currency'].fillna('Unknown')

# 6. Drop rows with missing ship-city/state/postal-code/country (only 33 rows)
df = df.dropna(subset=['ship-city', 'ship-state', 'ship-postal-code', 'ship-country'])

# Verify
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace('-','_')
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'sales_channel_':'sales_channel'})
```

```python
df.columns
```

```python
# 1. Drop rows with missing critical fields
df.dropna(subset=['order_id', 'amount', 'category'], inplace=True)

# 2. Parse date column
df['date'] = pd.to_datetime(df['date'], format='%m-%d-%y', errors='coerce')

# 3. Remove duplicate orders
df.drop_duplicates(subset='order_id', inplace=True)

# 4. Standardise text columns
df['status'] = df['status'].str.strip().str.title()
df['fulfilment'] = df['fulfilment'].str.strip()
```

```python
df.isnull().sum()
```

```python
# Summary statistics
print(df.describe())

# Value counts for categorical columns
print(df['category'].value_counts())
print(df['status'].value_counts())
print(df['fulfilment'].value_counts())

# Correlation between numeric columns
print(df[['qty', 'amount']].corr())

# Detect outliers in Amount
Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['amount'] < Q1 - 1.5*IQR) | (df['amount'] > Q3 + 1.5*IQR)]
print(f"Outliers found: {len(outliers)}")
```

```python
# Standardise additional text columns from the 23-column dataset
df['style'] = df['style'].astype(str).str.strip()
df['sku'] = df['sku'].astype(str).str.strip()
df['asin'] = df['asin'].astype(str).str.strip()
df['sales_channel'] = df['sales_channel'].astype(str).str.strip()
df['ship_state'] = df['ship_state'].astype(str).str.strip().str.title()
df['ship_city'] = df['ship_city'].astype(str).str.strip().str.title()
```

```python
df['sales_channel'].value_counts()
df['sku'].value_counts().head()
df['asin'].value_counts().head()
df['style'].value_counts().head()
df['ship_state'].value_counts().head()
```

```python
# Time-based features
df['month'] = df['date'].dt.month_name()
df['week'] = df['date'].dt.isocalendar().week
df['day_of_week'] = df['date'].dt.day_name()

# Revenue segmentation
df['revenue_band'] = pd.cut(
    df['amount'],
    bins=[0, 500, 2000, 5000, np.inf],
    labels=['Low', 'Medium', 'High', 'Premium']
)

# Fulfilment flag
df['is_amazon_fulfilled'] = (
    df['fulfilment'].str.contains('Amazon', case=False)
)

# B2B flag (present in the 23-column dataset)
if 'b2b' in df.columns:
    df['is_b2b'] = df['b2b'].astype(bool)
 
# Revenue per unit (Amount / Qty), guarding against divide-by-zero
df['revenue_per_unit'] = np.where(df['qty'] > 0, df['amount'] / df['qty'], np.nan)

# Promotion flag
df['has_promotion'] = df['promotion_ids'] != 'No Promotion'
 
# Extract base SKU prefix (often encodes product line/category)
df['sku_prefix'] = df['sku'].str.split('-').str[0]
```
